## Investigation Overview


> The overall goal of this preentation is to figure out what features are best for predicting students' performance in an exam. The analysis revealed that students of parents with low education who completed a test preparation course performed better than students of parents with high education who did not undergo a test preparation course. Also, students of parents with low education who had standard lunch performed better than students of parents with high education who had free or reduced lunch. This indicates that both test preparation course and lunch have stronger impact on students' performance than parental level of eduction.

## Dataset Overview

> This dataset consists of the marks secured by about 1000 students in three subjects - math, readng and writing. There are 999 students' exam scores in three subjects (math, reading and writing). The dataset contains 999 observations and 9 variables. The variables include: 'gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course', 'math score', 'reading score', 'writing score' and 'average_score'. There are four numrical variables (which include 'math score', 'reading score', 'writing score' and 'average_score') and five categorical variables (which include 'gender', 'race/ethnicity', 'parental level of education', 'lunch', and 'test preparation course') in the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
%matplotlib inline
# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('new_exam_data.csv')
df.head()
```

```python
# defining function
def multivariate_violin(numerical_var_in_quotes, categorical_var_in_qoutes, categorical_hue_in_quotes):
    rcParams['figure.figsize']= 12,6
    yticks = np.arange(10, 110, 10)
    ylabel = [f'{i}' for i in yticks]
    sns.violinplot(data=df, x=categorical_var_in_qoutes, y=numerical_var_in_quotes, hue=categorical_hue_in_quotes, inner='quartile', palette=['#bc5090','#10485288'], split=True)
    plt.xlabel(categorical_var_in_qoutes, fontsize=14)
    plt.ylabel(numerical_var_in_quotes, fontsize=14)
    plt.yticks(yticks, ylabel)
    plt.title(f'Violin plot of {numerical_var_in_quotes} of {categorical_var_in_qoutes} based on their {categorical_hue_in_quotes} status'.upper(), fontsize=16, y=1.05)
    plt.show();
```

## Visualization 1
>Students who had standard lunch but did not undergo a test preparation course performed better than students who completed a test preparation course but had free or reduced lunch.

```python
multivariate_violin('average score', 'lunch', 'test preparation course')
```

From the diagram above, one can notice that the median score of students who had standard lunch but dut dd not complete a test preparation course is higher than the median score of students who free/reduced lunch but completed a test preparation course. This shows that lunch had a greater effect on students' performance than a test preparation course as shown above.

## Visualization 2

> Students of parents with low education who had standard lunch performed better than students of parents with high education who had free or reduced lunch.

```python
multivariate_violin('average score', "parents' educational status", 'lunch')
```

It can be observed from the above dagram that the median score of students who had standard lunch but whose parents had low education is higher than the median score of students who had free/reduced lunch and whose parents had high education. This shows that lunch had a greater effect on students' performance than parental level of educaton.

## Visualization 3

>Students of parents with low education who completed a preparation course performed better than students of parents with high education who did not undergo a test preparation course.

```python
multivariate_violin('average score', "parents' educational status", 'test preparation course')
```

The above dagram shows that the median score of students who completed a test preparaton course but whose parents had low education is higher than the median score of students who dd not complete a test preparation course but whose parents had high education. This shows that test preparation course had a greater effect on students' performance than parental level of educaton.
